```python
# Ecuación diferencial de tasa de interés
sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
```

```python
S=sp.Function('S')
ode=sp.Eq(sp.Derivative(S(t),t)-1*S(t),0) # Sustituyendo c=1
ode
```

```python
sol=sp.dsolve(ode,S(t))  #Solución general
sol
```

```python
sol=sp.dsolve(ode,S(t),ics={S(0):10000})  #Condiciones iniciales
sol
```

```python
sp.plotting.plot(sol.rhs)
```

```python
constants={c:0.05}
sol.subs(constants)
```
