```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def base_df(ticker,inicio,intervalo):
  base=yf.Ticker(ticker)
  df=base.history(start=inicio,interval=intervalo)
  return df['Close']
```

```python
tesla=yf.Ticker('TSLA')
df=tesla.history(start='2023-12-13', interval='1d')
df.head(10)
```

```python
tesla_cierre=df['Close']
tesla_cierre.head(10)
```

```python
def serie_hist(datos,titulo,titulo_x,titulo_y):
    plt.figure(figsize=[12,8])
    plt.plot(datos)
    plt.grid()
    plt.title(titulo)
    plt.xlabel(titulo_x)
    plt.ylabel(titulo_y)
    return
```

```python
def retornos(base,shift):
  retorno=np.log(base)-np.log(base.shift(shift))
  return retorno.dropna()
```

```python
tesla_cierre=serie_hist(tesla_cierre,'Tesla cierre historico','Fecha','Precio de cierre')
```

```python
df_tesla=base_df('TSLA','2023-12-13','1d')
df_tesla.head(10)
```

```python
tesla_retorno=np.log(df_tesla)-np.log(df_tesla.shift(1))
tesla_ret=tesla_retorno.dropna()
tesla_ret.head(10)
```

```python
df_uber=base_df('UBER','2023-12-13','1d')
serie_hist(df_uber,'UBER cierre historico','Fecha','Precio de cierre')
```

```python
df_samsung=base_df('SMSN.IL','2023-12-13','1d')
serie_hist(df_samsung,'SAMSUNG cierre historico','Fecha','Precio de cierre')
```

```python
df_apple=base_df('AAPL','2023-12-13','1d')
serie_hist(df_apple,'APPLE cierre historico','Fecha','Precio de cierre')
```

```python
tesla_retornos=retornos(df_tesla,1)
uber_retornos=retornos(df_uber,1)
samsung_retornos=retornos(df_samsung,1)
apple_retornos=retornos(df_apple,1)
```

```python
plt.figure(figsize=[12,8])
plt.plot(tesla_retornos, color='red',label="TESLA")
plt.plot(uber_retornos,color='black',label='UBER' )
plt.plot(samsung_retornos,color='blue', label='SAMSUNG')
plt.plot(apple_retornos, color='green', label='APPLE')
plt.legend()
plt.grid()

```

```python
plt.figure(figsize=[12,8])
sns.set_theme(style="darkgrid")
plt.hist(tesla_retornos, color=sns.color_palette("dark")[0],alpha = 0.7,label="TESLA")
plt.hist(uber_retornos,color=sns.color_palette("dark")[1],alpha = 0.7,label='UBER' )
plt.hist(samsung_retornos,color=sns.color_palette("dark")[2], alpha = 0.7, label='SAMSUNG')
plt.hist(apple_retornos, color=sns.color_palette("dark")[3], alpha = 0.7,label='APPLE')
plt.legend()
plt.grid()
```
