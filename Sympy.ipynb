```python
import numpy as np
import sympy as sp
sp.init_printing()
```

```python
S=sp.Symbol('S')
t=sp.Symbol('t')
c=sp.Symbol('c')
```

```python
sp.expand((t+4)**2)
```

```python
((t+4)**2).subs(t,2)
```

```python
sp.factor(t**2+6*t+5)
```

```python
sp.simplify(t**2+6*t**2+2*t+5*t+5)
```

```python
sp.plotting.plot(t*t)
```

```python
sp.plotting.plot(100*sp.exp(0.15*t))
```

```python
# Ecuación diferencial de tasa de interés
sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
```

```python
S=sp.Function('S')
ode=sp.Eq(sp.Derivative(S(t),t)-1*S(t),0) # Sustituyendo c=1
ode
```

```python
sol=sp.dsolve(ode,S(t))  #Solución general
sol
```

```python
sol=sp.dsolve(ode,S(t),ics={S(0):10000})  #Condiciones iniciales
sol
```

```python
sp.plotting.plot(sol.rhs)
```

```python
constants={c:0.05}
sol.subs(constants)
```
