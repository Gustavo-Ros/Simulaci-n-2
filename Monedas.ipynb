```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
color=['Azul', 'Verde','Rojo','Naranja']
casilla=0
simulaciones=50000
tiro=rnd.choices([0, 1, 2], [0.25, 0.5, 0.25],k=simulaciones)
probabilidades=[0,0,0,0]
t=1
for t in range(len(tiro)):
 for i in range(tiro[t]):
  casilla=casilla+1
  if casilla==4:
    casilla=0
 probabilidades[casilla]+=1

probabilidades = [x / simulaciones for x in probabilidades]
print(probabilidades)

plt.bar(color, probabilidades, color=['blue', 'green', 'red', 'orange'])
plt.xlabel('Sector')
plt.ylabel('Probabilidad')
plt.title('Probilidad de cada sector')
plt.show()
```
